```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("House Price India.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
a = df.iloc[:, 0:22]
b = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.3, random_state=100)
```

```python
mse_values = []
for value in a_train.columns:
    rf = RandomForestRegressor()
    rf.fit(a_train, b_train)
    y_pred = rf.predict(a_test)
    mse_values.append(mean_squared_error(b_test, y_pred))
```

```python
mse_values = pd.Series(mse_values)
```

```python
mse_values.index = a_train.columns
```

```python
mse_values.sort_values(ascending=False).plot.bar()
```
